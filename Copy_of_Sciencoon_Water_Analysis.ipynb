Plotting a barplot for the pH value for each location under observation.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'pH',data =df,color='green');
sns.lineplot(data=df['pH'],marker='o',ms=6,lw=5,color='red',label='pH');

plt.rcParams.update({'font.size':70})
plt.title('pH PLOT')
plt.xticks(rotation=90);
```

Plotting a barplot for the temperature value for each location under observation.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'Temperature(0c)',data =df,palette='muted');
sns.lineplot(data=df['Temperature(0c)'],color='k',marker='s',ms=8,lw=5,label='Temperatue Plot(0c)')


plt.title('Temperatue Plot(0c)')
plt.xticks(rotation=90);
```

*Plotting a barplot for the pressure value for each location under observation.*

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'Pressure(mmHg)',data =df,color='gold');
sns.lineplot(data=df['Pressure(mmHg)'],color='k',marker='o',ms=8,lw=5,label='Pressure(mmHg)')


plt.title('PRESSURE PLOT')
plt.xticks(rotation=90);
```

Plotting a barplot for the resistivity value for each location under observation.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'Resistivity (m.cm)',data =df,color='purple');
sns.lineplot(data=df['Resistivity (m.cm)'],color='red',marker='o',ms=8,lw=5,label='Resistivity (m.cm)')

plt.title('RESISTIVITY PLOT')
plt.xticks(rotation=90);
```

Plotting a barplot for the conductivity value for each location under observation.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'Conductivity(mohm.cm)',data =df,palette='deep');
sns.lineplot(data=df['Conductivity(mohm.cm)'],color='red',marker='o',ms=8,lw=5,label='Conductivity(mohm.cm)')


plt.title('Conductivity Plot')
plt.xticks(rotation=90);
```

Plotting a barplot for the total dissolved solids value for each location under observation.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'Tdsppm',data =df,palette='dark');
sns.lineplot(data=df['Tdsppm'],color='red',marker='o',ms=8,lw=5,label='Tdsppm')


plt.title('TOTAL DISSOLVED SOLID PLOT')
plt.xticks(rotation=90);
```

Plotting a barplot for the salinity value for each location under observation.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'LOCATION',y= 'Salinity',data =df,color='red');
sns.lineplot(data=df['Salinity'],color='black',marker='o',ms=8,lw=5,label='Salinity')


plt.title('Salinity Plot')
plt.xticks(rotation=90);
```

Plotting a barplot for the % dissolved oxygen value for each location under observation.

Plotting the pressure distribution across all locatios under consideration.

Plotting the temperature distribution across all locatios under consideration.
